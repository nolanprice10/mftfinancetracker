# Quant Lab Notebook 03: Strategy + Metrics

Run a deterministic momentum strategy and compute common performance metrics.

```python
import numpy as np
import pandas as pd

df = pd.read_csv('../datasets/equities_sp500_proxy/v1/data.csv')
df['ret_5d'] = df['close'].pct_change(5)
df['signal'] = (df['ret_5d'] >= 0.02).astype(int)
df['strategy_ret'] = df['close'].pct_change().fillna(0) * df['signal'].shift(1).fillna(0)
df['equity'] = 10000 * (1 + df['strategy_ret']).cumprod()
df.tail()
```

```python
daily_ret = df['strategy_ret'].dropna()
cum_return = (df['equity'].iloc[-1] / df['equity'].iloc[0] - 1) * 100
rolling_peak = df['equity'].cummax()
drawdown = (df['equity'] / rolling_peak - 1) * 100
vol = daily_ret.std(ddof=1) * np.sqrt(252) * 100
sharpe = 0 if daily_ret.std(ddof=1) == 0 else (daily_ret.mean() / daily_ret.std(ddof=1)) * np.sqrt(252)

{
  'cumulative_return_pct': round(cum_return, 2),
  'max_drawdown_pct': round(drawdown.min(), 2),
  'volatility_pct': round(vol, 2),
  'sharpe_ratio': round(sharpe, 2)
}
```
