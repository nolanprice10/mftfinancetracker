# Quant Lab Notebook 02: Cleaning and Preprocessing

This notebook demonstrates simple, reproducible preprocessing steps.

```python
import pandas as pd

df = pd.read_csv('../datasets/crypto_btc_daily/v1/data.csv')
df['date'] = pd.to_datetime(df['date'])
df = df.drop_duplicates(subset=['date']).sort_values('date').reset_index(drop=True)

# Fill small missing gaps using forward fill
df[['open', 'high', 'low', 'close']] = df[['open', 'high', 'low', 'close']].ffill()
df['volume'] = df['volume'].fillna(0)

df.head()
```

```python
summary = {
    'row_count': len(df),
    'missing_total': int(df.isna().sum().sum()),
    'start': str(df['date'].min().date()),
    'end': str(df['date'].max().date())
}
summary
```
