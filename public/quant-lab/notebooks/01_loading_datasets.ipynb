# Quant Lab Notebook 01: Loading Datasets

This notebook shows how to load a versioned Quant Lab dataset from CSV and JSON.

```python
import pandas as pd

csv_path = '../datasets/equities_sp500_proxy/v1/data.csv'
json_path = '../datasets/equities_sp500_proxy/v1/data.json'

df_csv = pd.read_csv(csv_path)
df_json = pd.read_json(json_path)

print('CSV rows:', len(df_csv))
print('JSON rows:', len(df_json))
df_csv.head()
```

```python
print('Columns:', list(df_csv.columns))
print('Date range:', df_csv['date'].min(), '->', df_csv['date'].max())
```
